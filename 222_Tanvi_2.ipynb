```python
'''
Draw image color histogram for each of these images.
'''
# image color histogram for each resized image
for path in filelist:
  img = Image.open(path)

  img_arr = np.array(img)
  hist,bins = np.histogram(img_arr.flatten(),bins=50,range=[0,255])
  plt.bar(bins[:-1],hist,width=np.diff(bins))

  plt.title('Color Histogram')
  plt.xlabel('Color Pixels')
  plt.ylabel('Frequency')
  plt.show()
```

The color histograms show different color pixels and their frequency about which pixel ranges more than other. Some images here show sudden changes in pixels(specific colours spread over large range) while others change gradually(more colours fused in).

# 5. Text PreProcessing

```python
df = pd.read_csv('/content/a2_text - a2_text.csv')
df.head(10)
```

```python
''' For the content attribute, find length of content, number of
punctuations, number of stopwords, and number of times @ appears in the
tweet text. 
'''
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords') 
```

```python
def punct(txt): #counts punctuation in txt
  count = sum(1 for ch in txt if ch in string.punctuation)
  return count
```

```python
def stopword(txt): #count no of stopwords in text foe english language in consideration
  stopwrd = stopwords.words('english')
  count = sum(1 for word in txt.lower().split() if word in stopwrd)
  return count
```

```python
def mentions(txt):
  return (txt.count('@'))
```

```python
df['length'] = df['content'].apply(len)
df['punct'] = df['content'].apply(punct)
df['stopwords'] = df['content'].apply(stopword)
df['mentions'] = df['content'].apply(mentions)

df.head(10)
```

```python
'''
Draw correlation plot and find if any correlation exists between the sentiment attribute with these four new columns
'''

columns=['length','punct','stopwords','mentions']
#sentiment is encoded to plot its correlation with different attributes 
enc_sentiment = pd.get_dummies(df['sentiment'],prefix='sentiment')

df = pd.concat([df,enc_sentiment],axis=1)

df.head()
```

```python
for col in columns:
  corr_matrix = df[[col]+enc_sentiment.columns.tolist()].corr()

  plt.figure(figsize=(8,6))
  sns.heatmap(corr_matrix,annot=True)
  plt.title(f"{col} vs Sentiment")
  plt.show()

  correlations = corr_matrix[col][1:]
  print(f"Correlation value for {col} and sentiment: {correlations}")
```

```python
'''
Remove stopwords, and then tokenize words in each row of the ‘content’, count
the number of words, and save into a new attribute ‘wordCount’. Use the NLTK
library.
'''
from nltk.tokenize import word_tokenize
nltk.download('punkt')

stopwords = stopwords.words('english')
```

```python
def tokenize_after_stopwords_removed(txt):
  tokens = word_tokenize(txt)
  # filtered out stopwords from content attribute recieved in txt
  removed_tokens = [word for word in tokens if word.lower() not in stopwords]
  return len(removed_tokens)
```

```python
df['wordCount'] = df['content'].apply(tokenize_after_stopwords_removed)
df.head(10)
```

```python
'''
Create BoW (Bag of Words) vectors for each tweet content. Using
countvectorizer in the sklearn library.
'''
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
```

```python
#fit content attribute
bow_vect = vect.fit_transform(df['content'])
#text is transformed into BoW vector

#then convert BoW vector into dataframe 
bow_df = pd.DataFrame(bow_vect.toarray(), columns=vect.get_feature_names_out())
print(bow_df)
```

```python
'''
Draw separate word clouds for each of the following sentiment values: sadness,
enthusiasm, and neutral.
'''
from wordcloud import WordCloud

def generate_word_cloud(df,sentiment_value):
  sentiment_data = df[df['sentiment'] == sentiment_value]

  text = ' '.join(sentiment_data['content'])

  wordcloud = WordCloud().generate(text)

  plt.imshow(wordcloud, interpolation='bilinear')
  plt.title(f'{sentiment_value} Word Cloud')
  plt.axis('off')
  plt.show()
```

```python
generate_word_cloud(df,'sadness')
generate_word_cloud(df,'enthusiasm')
generate_word_cloud(df,'neutral')
```

Word Cloud helps in visualising the frequent usage and importance of different words in the text under consideration. We can find out state of user while commenting and their usual mood based on the frequently used words. WordCloud is of major use in sentiment analysis technique.
